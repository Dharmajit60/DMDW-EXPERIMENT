```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/DataSets/student-mat.csv")
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
columns_of_interest1 = ['age', 'G1', 'G2', 'G3', 'studytime', 'failures','traveltime','freetime']
```

```python
columns_of_interest2 = [ 'studytime','freetime']
```

```python
subset_data = df[columns_of_interest2]
```

```python
correlation_matrix = subset_data.corr()
```

```python
covariance_matrix = subset_data.cov()
```

```python
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix,cmap ="YlGnBu")
plt.show()
```

```python
print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
sns.heatmap(covariance_matrix,cmap ="coolwarm")
plt.show()
```
