```python
import numpy as np
import pandas as pd
import math
from scipy import stats

```

```python
n=int(input("Enter The Number Of Elements:"))
ls=[]
```

```python
for i in range(0,n):
  temp=float(input())
  ls.append(temp)
```

```python
sum=0
for i in range(0,n):
  sum=sum+ls[i]
```

```python
mean=sum/n
print("Mean: ",mean)
```

```python
ls.sort()
```

```python
sum2=ls[int(n/2)]+ls[int(n/2)-1]
median=sum2/2
print(median)
```

```python
counter = 0
max= ls[0]
for i in ls:
    curr_frequency = ls.count(i)
    if(curr_frequency> counter):
        counter = curr_frequency
        max = i
print(max)
```

```python
sum3=0
for i in range(0,n):
  temp=(ls[i]-mean)**2
  sum3=sum3+temp
var=(1/(n))*sum3
print(var)
```

```python
sd=math.sqrt(var)
print(sd)
```
